```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
```

```python
file_path= r'C:\Users\a---f\Documents\desarrollo\dhamma\enero_abril_totales_cursos.csv'
```

```python
def clean_column_name(column_name):
    column_name = column_name.strip()  # elimina caracteres no imprimibles y espacios en blanco
    column_name = unicodedata.normalize('NFKD', column_name)  # normaliza la cadena de texto
    return column_name
```

```python
cursos = pd.read_csv(file_path, encoding="Utf-8", encoding_errors="replace",index_col=0)
for col in cursos:
    new_col_name = clean_column_name(col)
    cursos = cursos.rename(columns={col: new_col_name})
cursos.columns = [col.replace("  ", "-") for col in cursos.columns]
    
cursos
```

```python
sns.barplot(x=cursos['Tipo de curso'], y=cursos['Waitlisted'])
```
