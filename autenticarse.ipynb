```python
import glob 
import time
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from getpass import getpass
```

```python
# Configuración del navegador (Firefox)
options = webdriver.FirefoxOptions()
options.headless = True
driver = webdriver.Firefox(options=options)

# Limpiar el navegador antes de agregar cookies
driver.delete_all_cookies()

# Autenticación en el sitio web 
login_url = 'https://latin.dhammareg.dhamma.org:8443/site/login'
username = input('Username')
password = getpass('Contraseña')

# Acceder a la página de inicio de sesión
driver.get(login_url)

# Autenticarse directamente en Selenium
driver.find_element(By.ID, 'LoginForm_username').send_keys(username)
driver.find_element(By.ID, 'passwordFieldClone0').send_keys(password)
driver.find_element(By.ID, 'login-button').click()

# Verificar si la autenticación fue exitosa
if "Totales de todos los cursos" in driver.title:  # Verificar si se redirige a la página de inicio después de iniciar sesión
    print("¡Autenticación exitosa!")
    

    # Acceder a la página donde se encuentra el elemento select después de la autenticación
    driver.get('https://latin.dhammareg.dhamma.org:8443/enrollments/admin?Enrollments%5BLocationId%5D=COM&Enrollments%5BCourseId%5D=')

    # desplegar menu cursos
    seleccionar_curso = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "CourseID")))

    # Crear objeto Select para interctuar con el elemento seleccion
    seleccion = Select(seleccionar_curso)

    nombre_curso = input('Ingrese el nombre del curso: ')
    
    seleccion.select_by_value(nombre_curso)
    print('Curso seleccionado correctamente')

    #Hacer clic en el boton buscar
    boton_buscar = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID, "search-button")))
    boton_buscar.click()
    print('Curso seleccionado correctamente y buscando')


    # Localizar el elemento select por ID
    select_element = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.ID, "columnSelector")))
        
    # Crear objeto Select para interactuar con el elemento select
    select = Select(select_element)

    # Seleccionar la opción con valor "6"
    select.select_by_value("6")
        
    print("Opcion Ver todo expandido seleccionada exitosamente.")

    time.sleep(10)

    # Esperar a que aparezca el enlace de exportar inscripciones y hacer clic en él
    enlace_exportar = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "ExportEnrollmentsLink")))
    enlace_exportar.click()
        
    print("Exportando inscripciones.")
    
    # Esperar un tiempo para que se complete la descarga
    time.sleep(10) 

    # Ruta de la carpeta donde se descargan los archivos
    carpeta_descargas = r'C:\Users\a---f\Downloads'

    # Obtener la lista de archivos xls en la carpeta de descargas
    archivos_xlsx = glob.glob(os.path.join(carpeta_descargas, '*.xlsx'))

    # Si no hay archivos xls, mostrar un mensaje y salir
    if not archivos_xlsx:
        print("No se encontraron archivos con la extensión .xls en la carpeta de descargas.")
        exit()

    # Obtener el archivo más reciente
    archivo_mas_reciente = max(archivos_xlsx, key=os.path.getctime)
    
    # Esperar un tiempo para que se complete el proceso
    time.sleep(10)

    # Mover el archivo descargado a la carpeta deseada
    nombre_archivo = 'lista.xlsx'
    ruta_destino = os.path.join(os.path.expanduser('~'), 'Documents\\desarrollo\\dhamma', nombre_archivo)
    os.rename(archivo_mas_reciente, ruta_destino)


    print(f"El archivo se ha descargado en la carpeta 'Documentos' como '{nombre_archivo}'")

    # Cerrar el navegador
    driver.quit()

else:
    print("Error de autenticación")
```
